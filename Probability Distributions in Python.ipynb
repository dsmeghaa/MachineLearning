normal.png

```python
#visualization
from scipy.stats import norm

# Parameters
mu = 800  # Mean
sigma = 200  # Standard deviation

# Create the distribution
x = range(0,1601)  # Adjust range as needed
probs = norm.pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, probs)
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution (μ={}, σ={})'.format(mu, sigma))
plt.show()
```

```python
#Solving Problems (Example: Probability of a value between 1000 and 500):
prob_between_minus1_and_1 = norm.cdf(1000, mu, sigma) - norm.cdf(500, mu, sigma)
print(f"Probability of a value between -1 and 1: {prob_between_minus1_and_1}")
```

##3. Uniform Distribution

uniform.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform
```

```python
def visualize(a, b, range_of_interest=None):
    # Generate values
    x = np.linspace(a - 5, b + 5, 1000)
    pdf = uniform.pdf(x, loc=a, scale=b-a)

    # Plot the distribution
    plt.figure(figsize=(8, 4))
    plt.plot(x, pdf, label=f"Uniform({a}, {b})", color='blue')
    plt.fill_between(x, pdf, where=((x >= (range_of_interest[0] if range_of_interest else a)) &
                                     (x <= (range_of_interest[1] if range_of_interest else b))),
                     color='skyblue', alpha=0.5, label="Area of interest")
    plt.title(f"Uniform Distribution")
    plt.xlabel("x")
    plt.ylabel("Density")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
# Function to calculate and print the probability for a range
def calculate_probability(a, b, lower, upper=None):
    if upper is None:  # Case for "greater than" probability
        prob = 1 - uniform.cdf(lower, loc=a, scale=b - a)
    else:
        prob = uniform.cdf(upper, loc=a, scale=b - a) - uniform.cdf(lower, loc=a, scale=b - a)
    return prob
```

#Identifying Which Distribution to Use

##Discrete vs Continuous

* Discrete distributions model events with a finite number of outcomes (e.g., number of successes in trials, number of accidents).
* Continuous distributions model variables that can take any value within a range (e.g., time, weight, height).

##Type of Problem

* If you're counting successes in trials, it's likely binomial.
* If you're counting events over fixed intervals (like time), it's likely Poisson.
* If you're measuring the time to the first success, consider geometric.
* If you want to model waiting time or time between events (e.g., call arrivals), consider exponential or gamma.
* If all outcomes within a range are equally likely, consider uniform.
